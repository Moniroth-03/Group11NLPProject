1. Load and Preprocess Corpus

```python
# Read and preprocess the text
with open('Machine_Learning.txt', 'r', encoding='utf-8') as file:
    text = file.read().replace('\n', '')

print(text)

# Tokenizing text into Sentences
sentences = sent_tokenize(text)

# Split into training, validation, and test sets
train_size = int(0.7 * len(sentences))
val_size = int(0.1 * len(sentences))

train = sentences[:train_size]
valid = sentences[train_size:train_size + val_size]
test = sentences[train_size + val_size:]

# Print the sizes of the datasets
print("Training set size:", len(train))
print("Validation set size:", len(valid))
print("Test set size:", len(test))
print("Total sentences:", len(sentences))
```

```python
# Tokenizing the Training Set into Words
all_text = ' '.join(train)
tokens = word_tokenize(all_text)

# Display tokenized words and their count
print(tokens)
print("Total tokens:", len(tokens))
```

2. Limit Vocabulary and Replace OOV(Out Of Vocabulary) word

```python
# Create vocabulary from the most common tokens
vocab_size = 1000
word_freq = Counter(tokens).most_common(vocab_size)

vocab = [word for word, _ in word_freq]

print("Vocabulary size:", len(word_freq))
print("Vocabulary:", vocab)
```

```python
# Replace tokens not in vocabulary with <UNK>
train_tokens = [word if word in vocab else "<UNK>" for word in tokens]

print("Train tokens:", train_tokens)
print(f"Number of <UNK> tokens in train_tokens: {train_tokens.count('<UNK>')}")
```

```python
def build_ngram_counts(tokens, n):
    """Build n-gram counts from tokenized data."""
    ngram_counts = defaultdict(int)
    for i in range(len(tokens) - n + 1): # Number of possible n-gram = L - n + 1, n size of n-gram
        
        ngram = tuple(tokens[i:i + n]) # Tuples are used to store multiple items in a single variable.
        ngram_counts[ngram] += 1  # list[start:stop]
    return ngram_counts # Output: {('the',): 2, ('cat',): 1, ('in',): 1, ('hat',): 1}

# Build counts for unigrams, bigrams, trigrams, and 4-grams
unigram_counts = build_ngram_counts(train_tokens, 1)
bigram_counts = build_ngram_counts(train_tokens, 2)
trigram_counts = build_ngram_counts(train_tokens, 3)
fourgram_counts = build_ngram_counts(train_tokens, 4)

# Print n-gram counts
print("Unigram Counts:", unigram_counts)
print("Bigram Counts:", bigram_counts)
print("Trigram Counts:", trigram_counts)
print("Fourgram Counts:", fourgram_counts)
```

4. Define Backoff Model

```python
#LM1: Backoff Model
class BackoffModel():
    def __init__(self, train_tokens, ngram_size=4):
        """Initialize the backoff model with token data and build n-gram counts."""
        self.train_tokens = train_tokens
        self.ngram_size = ngram_size
        self.ngram_counts = {i: build_ngram_counts(train_tokens, i) for i in range(1, ngram_size + 1)}
    
    def backoff_probability(self, ngram): # ngram = ("1", "2", "3", ...)
        #Calculate the backoff probability for a given n-gram.
        n = len(ngram)
        for i in range(n, 0, -1): # Iterate from full n-gram size to lower n-gram levels
            if ngram in self.ngram_counts[i]: # Check if the n-gram exists
                count_ngram = self.ngram_counts[i][ngram]
                count_total = sum(self.ngram_counts[i].values())
                return count_ngram / count_total # Return probability based on frequency counts
        return 1e-6 #  # Return a very small probability for unseen n-grams
    
    
    def calculate_perplexity(self, test_tokens):
        """Compute the perplexity of a test set given the backoff model."""
        log_prob_sum = 0
        for i in range(len(test_tokens) - self.ngram_size + 1): # [0, 1, ... n-1], (+ 1)
            ngram = tuple(test_tokens[i:i + self.ngram_size])  # Extract n-gram from test tokens
            prob = self.backoff_probability(ngram)
            log_prob_sum += math.log(prob) # Accumulate log probabilities
        perplexity = math.exp(-log_prob_sum / len(test_tokens))  # # Calculate the final perplexity value PP(W) = e^(sum(log(P(w))))
        return perplexity
```

5. Define Interpolation with Add-k Smothing

```python
# LM2: Interpolation Model with Add-k Smoothing
class InterpolationModel:
    def __init__(self, train_tokens, ngram_size=4, k=1, lambdas=None):
        """Initialize the interpolation model with token data, n-gram counts, and smoothing parameters."""
        self.train_tokens = train_tokens
        self.ngram_size = ngram_size
        self.k = k
        self.lamdas = lambdas if lambdas else [0.25, 0.25, 0.25, 0.25] # Default values for λ for 4-gram interpolation
        self.ngram_counts = {i: build_ngram_counts(train_tokens, i) for i in range(1, ngram_size + 1)}
        
    def interpolation_probability(self, ngram):
        """Calculate the probability of an n-gram using interpolation and add-k smoothing."""
        n = len(ngram)
        prob = 0
        for i in range(1, n + 1): # Iterate over levels of n-grams from unigram to full n-gram size
            ngram_count = self.ngram_counts[i].get(ngram, 0) + self.k # Add-k smoothing to the count C(ngram) + k
            
            total_count = sum(self.ngram_counts[i].values()) + self.k * len(self.ngram_counts[i]) # Add-k smoothing to total counts C(total) + k*v, v unique word
            
            prob += (self.lamdas[i-1] * (ngram_count / total_count)) # Weighted probability using lambda values PP(w) = λ_i * P(w_i)
        return prob
    
    def calculate_perplexity(self, test_tokens):
        """Compute the perplexity of a test set based on the interpolation model."""
        log_prob_sum = 0
        for i in range(len(test_tokens) - self.ngram_size + 1):
            ngram = tuple(test_tokens[i:i + self.ngram_size]) # Extract n-gram from the test tokens
            prob = self.interpolation_probability(ngram)
            log_prob_sum += math.log(prob) # Accumulate log probability
        perplexity = math.exp(-log_prob_sum / len(test_tokens))  # Compute final perplexity value
        return perplexity
```

6. Evaluate Models with Perplexity

```python
backoff_model = BackoffModel(train_tokens)
backoff_perplexity = backoff_model.calculate_perplexity(test)

print(f"LM1 Backoff Method(Computation is unsmoothed): {backoff_perplexity}")

# Train LM2: Interpolation Model with Add-k Smoothing
interpolation_model = InterpolationModel(train_tokens, k=1, lambdas=[0.2, 0.3, 0.3, 0.2])
interpolation_perplexity = interpolation_model.calculate_perplexity(test)

print(f"LM2 Interpolation Method(add-k smoothing): {interpolation_perplexity}")
```

7. Text Generation

```python
import random

"""
    Generate text using a given language model (Backoff or Interpolation).

    Args:
        model: The language model instance (BackoffModel or InterpolationModel).
        seed: List of initial tokens to start text generation.
        length: Number of tokens to generate.

    Returns:
        str: The generated text as a single string.
    """
def generate_text(model, seed, length=20):
    
     # Ensure the seed is a list of tokens
    generated = list(seed)
    
    for _ in range(length):
        # Determine context: Take the last n-1 tokens from the generated sequence. Get the last 3 tokens (for 4-gram, we need 3 previous tokens)
        context = tuple(generated[-(model.ngram_size-1):])# Extract the context window
        
         # Fetch possible next tokens and their corresponding probabilities
        possible_tokens = []
        probabilities = []
        
        # Iterate over ngrams in the highest ngram size's keys to find matches. Check all 4-grams that start with our context
        for ngram in model.ngram_counts[model.ngram_size-1].keys():
            if len(ngram) >= model.ngram_size-1 and ngram[:model.ngram_size-1] == context:
                next_token = ngram[-1]  # Get the last  token predicted by the matched ngram
                # Handle tuple vs string token. Ensure next_token is a string
                if isinstance(next_token, tuple):
                    next_token = next_token[-1]
                
                # Depending on model type, compute probability
                if isinstance(model, InterpolationModel):
                    prob = model.interpolation_probability(ngram) # Interpolation logic
                else:
                    prob = model.backoff_probability(ngram) # Backoff logic

                possible_tokens.append(next_token)
                probabilities.append(prob)
        
        # If no matching n-grams were found, fallback to unigram probabilities
        if not possible_tokens:
            possible_tokens = list(model.ngram_counts[1].keys())
            probabilities = [1/len(possible_tokens)] * len(possible_tokens) # Uniform probability for unigrams
        
        # Normalize probabilities
        total_prob = sum(probabilities)
        if total_prob > 0:
            probabilities = [p/total_prob for p in probabilities]
        
        # Randomly select the next token
        next_token = random.choices(possible_tokens, weights=probabilities)[0]

        # Handle tuple normalization if necessary. Ensure next_token is a string before appending
        if isinstance(next_token, tuple):
            next_token = next_token[-1]
            
        # Append the predicted token to the generated sequence
        generated.append(next_token)
    
    # Convert the token list into a single space-separated string and return
    return ' '.join(str(token) for token in generated)  # Convert all tokens to strings
```
